#### Filter out businesses in Madison

```python
f = open('yelp_dataset/yelp_academic_dataset_business.json', encoding='UTF-8')
business_dict = defaultdict(list)
entry = next(f)
while True:
    try: 
        entry = json.loads(entry)
        for key, value in entry.items():
            if key == 'city' and value=='Madison':
                for key, value in entry.items():
                    business_dict[key].append(value)
        entry = next(f)
    except:
        break
```

```python
business = pd.DataFrame(business_dict)
business
```

#### Categorize by Restaurants

```python
business = business[business.categories.str.contains('Restaurants', na=False)]
business_list = set(business['business_id'])
business
```

### Load Review data

```python
f = open('yelp_dataset/yelp_academic_dataset_review.json', encoding='UTF-8')
review_dict = defaultdict(list)
entry = next(f)
while True:
    try: 
        entry = json.loads(entry)
        for key, value in entry.items():
            if key == 'business_id':
                if value in business_list:
                    for key, value in entry.items():
                        review_dict[key].append(value)
        entry = next(f)
    except:
        break
```

```python
review = pd.DataFrame(review_dict)
user_list = set(review['user_id'])
review
```

### Load User data

```python
f = open('yelp_dataset/yelp_academic_dataset_user.json', encoding='UTF-8')
user_dict = defaultdict(list)
entry = next(f)

while True:
    try:
        entry = json.loads(entry)
        for key, value in entry.items():
            if key == 'user_id':
                if value in user_list:
                    for key, value in entry.items():
                        user_dict[key].append(value)
        entry = next(f)
    except:
        break
```

```python
user = pd.DataFrame(user_dict)
user
```

## Merge all dataframes

```python
data = pd.merge(review, business, on=['business_id'])
data = pd.merge(data, user, on=['user_id'])
data
```
