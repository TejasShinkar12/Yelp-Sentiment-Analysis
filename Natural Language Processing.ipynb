## Open CSV File

```python
import pandas as pd
import numpy as np
```

```python
data_cleaned = pd.read_csv('yelp_data_cleaned.csv', na_values = ['nan'])
data_cleaned
```

## Preprocess Review Data

```python
text_doc = data_cleaned['text']
text_doc_list = list(text_doc)
text_doc_test = text_doc_list[0:]
text_doc_test_2 = text_doc.iloc[0:]
```

### Import and clean text dataset

```python
# Run this cell ONLY ONCE, then comment it out

#import nltk
#nltk.download()
```

```python
from nltk.corpus import stopwords
from nltk.stem.wordnet import WordNetLemmatizer
import string

stop = set(stopwords.words('english'))
exclude = set(string.punctuation)
lemma = WordNetLemmatizer()
```

```python
def clean(doc):
    try:
        number_free = ''.join([i for i in doc if not i.isdigit()])
        stop_free = ' '.join([i for i in number_free.lower().split() if i not in stop and len(i)>2])
        punc_free = ''.join(ch for ch in stop_free if ch not in exclude)
        normalized = ' '.join(lemma.lemmatize(word) for word in punc_free.split())
    except:
        print(doc)
    return normalized
```

```python
doc_clean = [clean(doc).split() for doc in text_doc_test]
doc_clean
```

```python
# Importing Gensim
from gensim import *
from gensim.corpora import Dictionary

# creating the term dictionary for our corpus, where every unique term is assigned an index
dictionary = corpora.Dictionary(doc_clean)

# converting corpus into Document Term Matrix using using dictionary prepared above
doc_term_matrix = [dictionary.doc2bow(doc) for doc in doc_clean]
```

```python
import gensim
LDA = gensim.models.ldamodel.LdaModel

# Running and traininng LDA model on the Document Term Matrix
ldamodel = LDA(doc_term_matrix, num_topics = 50, id2word = dictionary, passes = 25)
```

```python
print(ldamodel.print_topics())
```

```python
import pyLDAvis
import pyLDAvis.gensim
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

pyLDAvis.enable_notebook()
vis = pyLDAvis.gensim.prepare(ldamodel, doc_term_matrix, dictionary)
vis
```

## Latent Semantic Search (LSA)

```python
# Preprocessing
text_doc = text_doc_test_2.str.replace("[^a-zA-Z#]"," ")

# removing short words
text_doc = text_doc.apply(lambda x: ' '.join([w for w in x.split() if len(w)>3]))

# make all text lowercase
text_doc = text_doc.apply(lambda x: x.lower())
```

```python
text_doc
```

```python
from nltk.corpus import stopwords
stop_words = stopwords.words('english')

# tokenization
tokenized_doc = text_doc.apply(lambda x: x.split())

# remove stop words
tokenized_doc = tokenized_doc.apply(lambda x: [item for item in x if item not in stop_words])

# de-tokenization
lemma = WordNetLemmatizer()
detokenized_doc = []

for i in range(len(text_doc_test_2)):
    t = ' '.join(tokenized_doc[i])
    t = ' '.join(lemma.lemmatize(word) for word in t.split())
    detokenized_doc.append(t)
    
text_doc = detokenized_doc
```

```python
text_doc
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer(stop_words='english', max_features=1000, max_df = 0.5, smooth_idf=True)
x= vectorizer.fit_transform(text_doc)
x.shape
```

```python
from sklearn.decomposition import TruncatedSVD

# SVD represent documents and terms in vectors
svd_model = TruncatedSVD(n_components=50, algorithm='randomized', n_iter=100, random_state=526)

svd_model.fit(x)

len(svd_model.components_)
```

```python
svd_model.components_
```

```python
terms = vectorizer.get_feature_names_out()

for i, comp in enumerate(svd_model.components_):
    terms_comp = zip(terms, comp)
    sorted_terms = sorted(terms_comp, key=lambda x: x[1], reverse=True)[:10]
    print("Topic "+str(i)+": ", end=' ')
    print(' '.join([x[0] for x in sorted_terms]))
```

## Analyze reviews by Low Rating

```python
data_lowrating = data_cleaned[data_cleaned['review_stars'] < 4]
```

```python
text_doc_low = data_lowrating['text']
```

```python
text_doc = text_doc_low.str.replace("[^a-zA-Z#]", " ")
text_doc = text_doc.apply(lambda x: ' '.join([w for w in x.split() if len(w) > 3]))
text_doc = text_doc.apply(lambda x: x.lower())
text_doc.index = np.arange(0, len(text_doc))
```

```python
from nltk.corpus import stopwords
stop_words = stopwords.words('english')
tokenized_doc = text_doc.apply(lambda x: x.split())
tokenized_doc = tokenized_doc.apply(lambda x: [item for item in x if item not in stop_words])
lemma = WordNetLemmatizer()

detokenized_doc = []
for i in range(len(text_doc_low)):
    t = ' '.join(tokenized_doc[i])
    t = ' '.join(lemma.lemmatize(word) for word in t.split())
    detokenized_doc.append(t)
    
text_doc = detokenized_doc
```
